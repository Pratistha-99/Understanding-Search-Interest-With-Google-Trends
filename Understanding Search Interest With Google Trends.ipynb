```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
trends = pd.read_csv('ai_tools_comparison.csv')

# Inspect the data
print(trends.head())
print(trends.info())
```

Identify the AI tool that has shown the most consistent growth in interest. 
Visualize the interest levels of ChatGPT, Gemini, and Microsoft Copilot over time.

```python
# Calculate the monthly growth
growth = trends.set_index('week').pct_change().fillna(0) 
```

```python
# Measure consistency of growth
consistency = growth.std()
most_consistent_tool = consistency.idxmin()  # Tool with the lowest standard deviation
print(f"The most consistent tool is {most_consistent_tool}")
```

```python
# Saving the data
most_consistent_tool = consistency.idxmin()
```

```python
tools = ['chatgpt', 'gemini', 'microsoft_copilot']

# Ensure that the column names in the DataFrame are stripped of any leading/trailing whitespace
trends.columns = trends.columns.str.strip()

# Check if all tools are present in the DataFrame columns
missing_tools = [tool for tool in tools if tool not in trends.columns]
if missing_tools:
    raise KeyError(f"Columns {missing_tools} not found in DataFrame")

trends_tools = trends[['week'] + tools]
```

```python
# Create Visualisation

plt.figure(figsize=(10, 6))
for tool in tools:
    plt.plot(trends['week'], trends[tool], label=tool)  # Plot each tool's interest over time
plt.xlabel('Week')
plt.ylabel('Interest Level')
plt.title('Interest Levels Over Time')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

Identify the month and year when ChatGPT experienced its largest decline in interest

```python
# Chatgpt decline
trends['chatgpt_diff'] = trends['chatgpt'].diff()
gpt_dip_date = trends.loc[trends['chatgpt'].idxmin(), 'week']  # Date with the largest decline
gpt_dip = pd.to_datetime(gpt_dip_date).strftime('%B %Y')  # Convert to "Month YYYY" format
print(f"The largest decline in interest for ChatGPT occurred in {gpt_dip}")
gpt_dip = pd.to_datetime(gpt_dip_date).strftime('%B %Y')
```

Analyze Seasonality and Monthly Averages

```python
trends['week'] = pd.to_datetime(trends['week']) #Convert Date column to datetime format
trends['Month'] = trends['week'].dt.strftime('%B') #Extract the month from the date
# Compute monthly average
monthly_avg = trends.groupby('Month').mean().mean(axis=1)
best_month = monthly_avg.idxmax()  
# Month with the highest average interest
print(f"The month with the highest average interest is {best_month}")
best_month = monthly_avg.idxmax()
```

Checking the result

```python
print(f"Most consistent tool: {most_consistent_tool}")
print(f"Largest decline in interest for ChatGPT: {gpt_dip}")
print(f"Month with highest average interest: {best_month}")
```
